## Exercise 16 — Find vs membership

Using `s = 'Neuroscience Rocks!'`:
- use `.find()` to locate `'science'`,
- search for `'Psychology'`,
- test whether `'Rocks'` is present with `in`,
- create a boolean that is `True` only if `'science'` was found.

Note: `.find()` returns the starting index if the substring is found, and `-1` otherwise.


```python
s = 'Neuroscience Rocks!'

idx_science = s.find('science')
idx_psy = s.find('Psychology')
has_rocks = 'Rocks' in s
science_found = idx_science != -1

print(idx_science)
print(idx_psy)
print(has_rocks)
print(science_found)
```

## Exercise 17 — Mini-analysis

A participant completed a task with `participant_id = 'sub-07'`, `n_correct = 47`, `n_total = 50`, and `mean_rt = 0.61`.

Compute accuracy, create `passed_accuracy`, `fast_enough`, and `include`, then build and print a summary string like `sub-07: accuracy=0.94, include=True`.

```python
participant_id = 'sub-07'
n_correct = 47
n_total = 50
mean_rt = 0.61

accuracy = n_correct / n_total
passed_accuracy = accuracy > 0.80
fast_enough = mean_rt < 0.70
include = passed_accuracy and fast_enough
summary = f"{participant_id}: accuracy={accuracy}, include={include}"

print(summary)
```

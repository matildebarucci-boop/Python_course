## Exercise 1 — Excluding bad EEG trials

You performed EEG recordings from a subject in 10 trials total. Some trials have been marked as `"BAD"` due to eye blinks and poor electrode contact. Create a new list containing only the good trials.

Questions:
- What container type should `good_trials` be?
- Why is a list appropriate here?

```python
trial_ids = ["001", "002", "003", "004", "005",
             "006", "007", "008", "009", "010"]
bad_trials = ["004", "006", "007"]

good_trials = [trial for trial in trial_ids if trial not in bad_trials]
print(good_trials)
```

## Exercise 2 — Trial filtering with two conditions

Starting from the same trial list, create a new list `usable_trials` that:
1. excludes all bad trials
2. excludes all trials with an odd trial number

The final result should only contain good **even-numbered** trials.

```python
trial_ids = ["001", "002", "003", "004", "005",
             "006", "007", "008", "009", "010"]
bad_trials = ["004", "006", "007"]

usable_trials = [
    trial for trial in trial_ids
    if trial not in bad_trials and int(trial) % 2 == 0 # % 2 calcola il resto della divisione per 2. Se il resto è 0, il numero è pari. Se il resto fosse 1, il numero sarebbe dispari.
]
print(usable_trials)
```

## Exercise 3 — Summary statistics on participant ages

You collected participant ages in a study.

Compute and print:
- number of participants
- minimum age
- maximum age
- mean age
- the number of times age `22` appears

Then create a new list `older_than_21` containing only ages strictly greater than 21.

```python
ages = [19, 22, 25, 21, 20, 24, 22, 19, 26, 23, 21]

print(len(ages))              # number of participants
print(min(ages))              # minimum age
print(max(ages))              # maximum age
print(sum(ages) / len(ages)) # mean age
print(ages.count(22))           # how many times 22 appears

older_than_21 = [ages for ages in ages if ages > 21]

print(older_than_21)
```

## Exercise 5 — append() vs extend() vs +

Create three separate lists starting from the same `base` list.

Create:
- `with_append` by adding `["temporal", "occipital"]` using `append()`
- `with_extend` by adding the same values using `extend()`
- `with_plus` by adding the same values using `+`

Questions:
- How does `with_append` differ from the other two? he added a list inside a list
- Which one contains a nested list? the with_append

```python
base = ["frontal", "parietal"]

with_append = base.copy()
with_append.append(["temporal", "occipital"])

with_extend = base.copy()
with_extend.extend(["temporal", "occipital"])

with_plus = base + ["temporal", "occipital"]

print(with_append)
print(with_extend)
print(with_plus)
```

## Exercise 8 — Comparing sets of artifact labels

Two raters marked artifact types in an EEG session.

Compute and print:
- labels identified by both raters
- all labels identified by at least one rater
- labels only in `rater1`
- labels only in `rater2`

```python
rater1 = {"blink", "muscle", "drift", "line_noise"}
rater2 = {"blink", "drift", "movement", "electrode"}

print(rater1 & rater2)  # both raters
print(rater1 | rater2)  # at least one rater
print(rater1 - rater2)  # only rater1
print(rater2 - rater1)  # only rater2
```

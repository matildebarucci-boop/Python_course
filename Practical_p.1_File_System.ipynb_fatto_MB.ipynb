## Exercise 4 — Count files by extension

Given a folder, count how many files end with:

- `.py`
- `.ipynb`
- `.md`

Print the three counts.

```python
from pathlib import Path

folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course")

py_count = 0
ipynb_count = 0
md_count = 0

for path in folder.glob("*"):
    if path.suffix == ".py":
        py_count += 1
    elif path.suffix == ".ipynb":
        ipynb_count += 1
    elif path.suffix == ".md":
        md_count += 1

print("py:", py_count)
print("ipynb:", ipynb_count)
print("md:", md_count)
```

## Exercise 6 — Recursive notebook search

Search recursively inside a course folder and print all `.ipynb` files found in all subfolders.

```python
from pathlib import Path

course_folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course")

count = 0

for path in course_folder.rglob("*.ipynb"):
    print(path)
    count += 1

print("Total notebooks:", count)
```

## Exercise 8 — Create a nested folder structure

Create the following structure inside a base folder:

## Exercise 9 — Move matching files

Inside a folder, move all `.txt` files into a subfolder called `text_files`.

If the destination folder does not exist, create it first.

```python
from pathlib import Path

folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course")
destination = folder / "exercise9_txt"

destination.mkdir(exist_ok=True)

for path in folder.glob("*txt"):
    if not path.is_dir():
        path.rename(destination / path.name)
```

## Exercise 10 — Organize files by prefix

Suppose a folder contains files with names such as:

- `sub01_session1_data.csv`
- `sub01_session2_data.csv`
- `sub02_session1_data.csv`

Write code that moves each file into a folder structure like:

- `sub01/session1/`
- `sub01/session2/`
- `sub02/session1/`

Use the filename (without extension) and split it on `"_"`. 

```python
from pathlib import Path

data_folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course\exercise10_csv")
new_root = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course\exercise10_csv_new")
new_root.mkdir(exist_ok=True)

for file in data_folder.glob("*.csv"):
    subject, session, _ = file.stem.split("_")
    target_folder = new_root / subject / session #nella cartella nuova adesso abbiamo una cartella per sogg con dentro le sessioni
    target_folder.mkdir(exist_ok=True, parents=True)
    file.replace(target_folder / file.name)
```

## Exercise 11 — Build a file inventory

Given a folder, create a list of dictionaries where each dictionary contains:

- `"name"`
- `"suffix"`
- `"is_dir"`
- `"parent"`

for every path in the folder.

Then print the resulting list.

```python
from pathlib import Path

folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course")
inventory = []

for path in folder.glob("*"):
    info = {
        "name": path.name,
        "suffix": path.suffix,
        "is_dir": path.is_dir(),
        "parent": str(path.parent)
    }
    inventory.append(info)

print(inventory)
```

## Exercise 12 —  Only files, sorted by suffix

Given a folder:

1. collect only the files, not the subfolders
2. sort them by suffix
3. print their names

```python
from pathlib import Path

folder = Path(r"C:\Users\matilde.barucci\Documents\GitHub\Python_course")

files_only = []

for path in folder.glob("*"):
    if not path.is_dir():
        files_only.append(path)

for path in sorted(files_only, key=lambda p: p.suffix):
    print(path.name)
```
